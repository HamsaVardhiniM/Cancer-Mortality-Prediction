## Handling null values

### for each col with null values, if more than 50 percent of the entire data in the col are null avlues, then the col is dropped, else the null values are filled with the mean of the col

```python
def handle_null_values(df):
    
    total_rows = len(df)
    for column in df.columns:
        null_count = df[column].isnull().sum()
        if null_count > 0:
            if null_count > (0.5 * total_rows):
                df.drop(columns=[column], inplace=True)
               
            else:
                mean_value = df[column].mean()
                df[column].fillna(mean_value, inplace=True)
    return df
```

```python
df = handle_null_values(df)
```

```python
null_vals=df.isnull().sum()
print(null_vals[null_vals>0])
```

## Checking for constant columns

### Constant column(a col in which all the values are same), it means that the col does not paly a role in the prediction of the target variable, in other words, it does not affect the target variable, so it can be removed

```python
def drop_constant_columns(df):
    for column in df.columns:
        if df[column].nunique() == 1:
            df.drop(columns=[column], inplace=True)
    return df
```

## Handling the outliers

## Finding the normally distributed cols using Shapiro- Wilk Test

```python
from scipy.stats import shapiro

def check_normal_distribution(df):
    normally_distributed_cols = []
    for column in df.select_dtypes(include=['number']).columns:
        stat, p_value = shapiro(df[column].dropna())
        if p_value > 0.05:
            normally_distributed_cols.append(column)
    
    return normally_distributed_cols
```

```python
normally_distributed_cols=check_normal_distribution(df)
print(normally_distributed_cols)
```

## Checking for outliers in normally_distributed_cols

```python
def find_outliers(df, normally_distributed_cols):
    outliers_present = []
    
    for column in normally_distributed_cols:
        mean_value = df[column].mean()
        std_dev = df[column].std()
        
        #  upper and lower limits for outliers
        upper_limit = mean_value + 3 * std_dev
        lower_limit = mean_value - 3 * std_dev
        outliers = df[(df[column] > upper_limit) | (df[column] < lower_limit)]
        if not outliers.empty:
            outliers_present.append(column)
            num_outliers = ((df[column] > upper_limit) | (df[column] < lower_limit)).sum()
            print( " number of outliers:",num_outliers)
    
    return outliers_present
```

```python
outliers_present = find_outliers(df, normally_distributed_cols)
print(outliers_present)
```

## Removing the outliers (Trimming)

```python
def remove_outliers(df, normally_distributed_cols):
    for column in normally_distributed_cols:
        mean_value = df[column].mean()
        std_dev = df[column].std()
        upper_limit = mean_value + 3 * std_dev
        lower_limit = mean_value - 3 * std_dev
        df = df[(df[column] <= upper_limit) & (df[column] >= lower_limit)]
    
    return df
```

```python
df=remove_outliers(df, normally_distributed_cols)
```

## Checking for skewed columns

```python
def find_skewed_columns(df):
    skewed_cols = []
    
    for column in df.select_dtypes(include=['number']).columns:
        skewness = df[column].skew()
        if abs(skewness) > 1:
            skewed_cols.append(column)
    
    return skewed_cols
```

```python
skewed_columns=find_skewed_columns(df)
print(skewed_columns)
```

```python
print(len(skewed_columns))
```

### Removing the outliers

```python
def find_outlier_columns(df, skewed_cols):
    outlier_cols = []

    for col in skewed_cols:
        if col in df.columns:
            Q1 = df[col].quantile(0.25)
            Q3 = df[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_range = Q1 - 1.5 * IQR
            upper_range = Q3 + 1.5 * IQR

            if ((df[col] < lower_range) | (df[col] > upper_range)).any():
                outlier_cols.append(col)

    return outlier_cols
```

```python
outlier_cols=find_outlier_columns(df, skewed_columns)
print(outlier_cols)
```

```python
print(len(outlier_cols))
```

## Capping the outliers

```python
def replace_outliers(df, skewed_cols):
    for col in skewed_cols:
        if col in df.columns:
            Q1 = df[col].quantile(0.25)
            Q3 = df[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_range = Q1 - 1.5 * IQR
            upper_range = Q3 + 1.5 * IQR

            df[col] = df[col].apply(lambda x: lower_range if x < lower_range else (upper_range if x > upper_range else x))
    
    return df
```

```python
df=replace_outliers(df, outlier_cols)
df.shape
```

```python
def box_plot(df, col):
    fig = go.Figure(data=[go.Box(y=df[col], boxpoints='outliers', jitter=0.3, pointpos=-1.8)])
    fig.update_layout(
        title='Boxplot of ' + col,
        yaxis_title=col,
        width=500,
        height=500)
    fig.show()

for col in outlier_cols :
    box_plot(df, col)
```
